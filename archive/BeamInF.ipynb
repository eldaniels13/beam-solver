<center>Condiciones de equilibrio</center>
$$\sum F_x=0$$
$$\sum F_y=0$$
$$\sum M=0$$

<center>Resolviendo el sistema</center>
$$A_x=0$$

$$A_y=\frac{ql}{2}$$

$$B_y=\frac{ql}{2}$$

<center>Por lo tanto, las ecuaciones V y M serán:</center>
$$M=A_x x-\frac{qx^2}{2}=\frac{q}{2}(lx-x^2)$$

$$V=A_x -qx = q(\frac{l}{2}-x)$$



```python
import numpy as np 
####Importamos la librería en la primera linea
####y lo ejecutamos al menos una vez
```

```python
#Ingresamos los datos iniciales (verificamos que sean unidades consistentes)
l=15 #longitud viga, m
q=35 #valor de la carga uniformemente distribuida, kN/m
```

```python
x=np.linspace(0,l,999) #creamos la cordenada x desde 0 a l
                        # n,valores a los que aplicaremos las ecuaciones M y V
M=q/2 * (l*x - x**2)    #Ecuación de momento flector
V=q *(l/2 - x)          #Ecuación de fuerza cortante
```

```python
from matplotlib import pyplot as plt    #Importamos la librería

#Ploteamos una líniea horizontal que regresente la longitud de la viga
axis=[0]*len(x)                      #creamos una variable llena de ceros con  igual número de elementos que x
plt.plot(x,axis,color='k')           #Ploteamos x,axis para visualizar la línea

#Ploteamos el momento flector
plt.plot(x,-M,label="M")             #Creamos el plot
plt.title('Diagramas V y M')         #Agregamos el título de la gráfica
plt.legend()                         #Activamos la leyenda en la gráfica
plt.xlabel('Longitud,m')             #Agregamos la etiqueta X
plt.ylabel('Momento, kN/m y Fuerza, kN')        #Agregamos la etiqueta Y
plt.rcParams['figure.figsize'] = [15, 8]   #Establecemos las dimensiones de la gráfica

Mmax=np.amax(M)                      #Buscamos el valor máximo dentro del array M
x_M = np.where(M == np.amax(Mmax))   #Localizamos la posición del valor máximo M dentro del array
M_x_coor=(x[x_M])                    #Ubicamos la posición en X correpondiente al valor (posición) Mmax

Mlabel="{:.2f}".format(Mmax)         #Creamos la etiqueta para el valor máximo con 2 decimales
plt.annotate(Mlabel,(M_x_coor,-Mmax), textcoords="offset points", # coordenadas de la etiqueta (x,y)
                xytext=(0,10), # offset desde las coordenadas (x,y))
                ha='center')
               


#Ploteamos la fuerza cortante
plt.plot(x,V, label="V")              #Creamos el plot
plt.legend()                          #Activamos la leyenda en la gráfica

Vmax=max(V)                           #Buscamos el valor máximo dentro del array V
x_V = np.where(V == np.amax(Vmax))    #Localizamos la posición del valor máximo V dentro del array
V_x_coor=(x[x_V])                     #Ubicamos la posición en X correpondiente al valor (posición) V

Vlabel="{:.2f}".format(Vmax)          #Creamos la etiqueta para el valor máximo con 2 decimales
plt.annotate(Vlabel,(V_x_coor,Vmax), textcoords="offset points", # coordenadas de la etiqueta (x,y)
                xytext=(0,-15),       # offset desde las coordenadas (x,y)
                 ha='center')         #centramos la etiqueta
                


Vmin=min(V)                           #Buscamos el valor mínimo dentro del array V
x_Vmin = np.where(V == np.amax(Vmin)) #Localizamos la posición del valor mínimo V dentro del array
V_xmin_coor=(x[x_Vmin])               #Ubicamos la posición en X correpondiente al valor (posición) V

Vminlabel="{:.2f}".format(Vmin)       #Creamos la etiqueta para el valor máximo con 2 decimales
plt.annotate(Vminlabel,(V_xmin_coor,Vmin), textcoords="offset points", # ubicación de los puntos (x,y)
                xytext=(0,+15),      # offset desde los puntos (x,y)
                ha='center')         #centramos la etiqueta
                 






```
